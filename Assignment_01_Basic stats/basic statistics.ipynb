```python
import matplotlib.pyplot as plt #Used for creating static, interactive, and animated plots
import seaborn as sns #A higher-level API built on top of matplotlib for attractive and informative statistical graphics
import pandas as pd # Used for data manipulation and analysis

# Optional: Set Seaborn style
sns.set(style="whitegrid") #Sets the Seaborn theme (background style)

# Create the 3 bar charts side by side
plt.figure(figsize=(18, 5)) #Creates a new figure (container for all subplots)

# 1️⃣ Bar Chart - BU
plt.subplot(1, 3, 1)  #Adds a subplot (a grid-based layout) to the figure
sns.countplot(x='BU', hue='BU', data=df, palette='Set2', legend=False)  #Sets the Seaborn theme (background style)
plt.title('Count of Business Units (BU)')  #Creates a new figure (container for all subplots)
plt.xlabel('BU')   #Sets the x-axis label
plt.ylabel('Count') #Sets the y-axis label

# 2️⃣ Bar Chart - Brand
plt.subplot(1, 3, 2)
sns.countplot(x='Brand', hue='Brand', data=df, palette='Set3', legend=False)
plt.title('Count of SKUs by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45) #Rotates x-axis tick labels by 45 degrees

# 3️⃣ Bar Chart - Top 10 Models
plt.subplot(1, 3, 3)
top_models = df['Model'].value_counts().head(10).index #Sets the Seaborn theme (background style)  #Selects the top 10 most frequent models
filtered_df = df[df['Model'].isin(top_models)]
sns.countplot(x='Model', hue='Model', data=filtered_df, palette='husl', legend=False)
plt.title('Top 10 Most Frequent Models')
plt.xlabel('Model')
plt.ylabel('Count')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# Standardization of Numarical Variables

```python
print(df.info()) #load dataset
print(df.head()) 
```

```python
# Select only numerical columns
numeric_cols = df.select_dtypes(include='number').columns

# Apply Z-score normalization
standardized_df = (df[numeric_cols] - df[numeric_cols].mean()) / df[numeric_cols].std()

# Add '_z' suffix to standardized columns
standardized_df.columns = [col + '_z' for col in numeric_cols]

# Combine with original DataFrame
df_standardized = pd.concat([df, standardized_df], axis=1)

# View result
print(df_standardized.head())
```

```python
plt.figure(figsize=(12, 5))

# Before standardization
plt.subplot(1, 2, 1)
sns.histplot(df['Net Sales Value'], kde=True, color='skyblue')
plt.title('Original Net Sales Value')

# After standardization
plt.subplot(1, 2, 2)
sns.histplot(df_standardized['Net Sales Value_z'], kde=True, color='orange')
plt.title('Standardized Net Sales Value (Z-score)')

plt.tight_layout()
plt.show()
```

```python
# Automatically select non-numeric columns (object or category dtype)
categorical_cols = df.select_dtypes(include='object').columns
print("Categorical Columns:\n", categorical_cols.tolist())
```

```python
# Apply one-hot encoding using pandas get_dummies
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=False)  # drop_first=True to avoid multicollinearity

# Show portion of transformed dataset
print(df_encoded.head())
```

```python
import pandas as pd

# Load your dataset
df = pd.read_csv("sales_data_with_discounts.csv")  # replace with actual file if needed

# Display original data shape
print("Original Data Shape:", df.shape)

# Apply One-Hot Encoding to all categorical columns
categorical_cols = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']
df_encoded = pd.get_dummies(df, columns=categorical_cols)

# Display transformed dataset shape and preview
print("Transformed Data Shape:", df_encoded.shape)
print(df_encoded.head())
```

# Explantation


Descriptive Analytics:

Calculated mean, median, mode, and standard deviation for numerical columns such as Volume, Avg Price, and Sales.

These metrics help understand the central tendency and variability of data.

Data Visualization:

Histograms were used to analyze the distribution and skewness of numerical columns.

Boxplots identified outliers and helped understand the spread using interquartile range.

Bar charts visualized the frequency of categories in columns like Brand, BU, and Model, giving insights into data composition.

Standardization:

Applied z-score normalization to numerical columns to bring all values to a common scale.

This step ensures that features with different units or ranges do not dominate others in analysis or ML models.

Categorical Data Encoding:

Converted categorical variables into dummy variables (one-hot encoding).

This makes categorical data suitable for machine learning algorithms by creating binary columns for each category.

Conclusion:

Descriptive analytics revealed trends and patterns in the sales data.

Data preprocessing steps like standardization and one-hot encoding are crucial to prepare the dataset for accurate and efficient machine learning and statistical modeling.

# a. 99% Confidence Interval

```python
import numpy as np 
import scipy.stats as stats
```

```python
# Data 
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
# Calculate Mean and Standard Deviation 
mean = np.mean(data)
std_dev = np.std(data,ddof=1)
n = len(data)
print("Sample Mean:", mean)
print("Sample Standard Deviation:", std_dev)
print("Sample Size:", n)
```

```python
# Find T-Criticsl valuer for 99% confidence 
confidence = 0.99
alpha = 1 - confidence 
df = n-1
t_critical = stats.t.ppf(1 - alpha/2,df)
margin_error = t_critical * (std_dev / np.sqrt(n))
lower = mean - margin_error
upper = mean + margin_error 
print("Confidence Interval(99%):",(lower,upper))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6, 4))
plt.errorbar(x=1, y=mean, yerr=margin_error, fmt='o', color='blue', capsize=10)
plt.title("99% Confidence Interval")
plt.xlim(0.5, 1.5)
plt.grid(True)
plt.show()
```

# b. Using Population Std_dev

```python
from scipy.stats import norm
```

```python
#population Std_dev
pop_std_dev = 0.2
confidence = 0.99
alpha = 1 - confidence 
```

```python
# Z-critical value for 99% confidence level
z_critical = norm.ppf(1 - alpha/2)
margin_error_z = z_critical * (pop_std_dev / np.sqrt(n))
```

```python
# Confidence interval
lower_z = mean - margin_error_z
upper_z = mean + margin_error_z
```

```python
print("Mean:", mean)
print("Z-Critical Value:", z_critical)
print("Margin of Error:", margin_error_z)
print("Confidence Interval using Population Std Dev (99%):", (lower_z, upper_z))
```

```python
plt.figure(figsize=(6, 4))
plt.errorbar(x=1, y=mean, yerr=margin_error_z, fmt='o', color='green', capsize=10)
plt.title("99% Confidence Interval using Population Std Dev")
plt.xlim(0.5, 1.5)
plt.grid(True)
plt.show()
```
